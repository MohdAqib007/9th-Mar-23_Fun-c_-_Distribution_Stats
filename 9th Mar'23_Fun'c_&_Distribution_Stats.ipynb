```python
#Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?
#Ans-
'''A Cumulative Density Function (CDF) is a function that gives the probability that a random variable takes a value less than or equal to a given value. 
It is defined for both continuous and discrete random variables and is often denoted by F(x).

For a continuous random variable, the CDF is given by the integral of the PDF from negative infinity to x:

F(x) = P(X ≤ x) = ∫-∞^x f(t) dt

For a discrete random variable, the CDF is the sum of the probabilities of all values less than or equal to x:

F(x) = P(X ≤ x) = ∑k≤x P(X=k)

The CDF is useful because it can be used to find probabilities for ranges of values of the random variable. 
For example, the probability that a continuous random variable X takes a value between a and b is given by:

P(a < X ≤ b) = F(b) - F(a)

Similarly, the probability that a discrete random variable X takes a value between a and b is given by:

P(a < X ≤ b) = F(b) - F(a-1)

As an example, consider the random variable X that represents the number of children in a family. Suppose that the PMF of X is given by:

P(X=0) = 0.1
P(X=1) = 0.4
P(X=2) = 0.3
P(X=3) = 0.2

The CDF of X is given by:

F(x) = P(X ≤ x) =

0 if x < 0
0.1 if 0 ≤ x < 1
0.5 if 1 ≤ x < 2
0.8 if 2 ≤ x < 3
1 if x ≥ 3
For example, the probability that a family has at least 2 children is given by:

P(X ≥ 2) = 1 - P(X < 2) = 1 - F(1) = 1 - 0.5 = 0.5

The CDF is an important tool in probability theory and statistics, as it allows us to find probabilities for ranges of values of a random variable. 
It is also used in hypothesis testing, where it is used to calculate p-values.'''
```

```python
#Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?
#Ans-
'''The Bernoulli distribution is a discrete probability distribution that describes the outcomes of a single binary experiment that can either have a success or a failure outcome. It is named after the Swiss mathematician Jacob Bernoulli who used it to study the probability of coin tosses.

The Bernoulli distribution has only one parameter, p, which represents the probability of success. The probability mass function (PMF) of the Bernoulli distribution is given by:

P(X=k) = p^k (1-p)^(1-k), for k = 0 or 1

where X is the random variable that takes on the value 1 if the experiment is a success and 0 if it is a failure.

Example: Suppose we toss a fair coin. The outcome of the experiment is either heads or tails. If we define success as getting heads, then the Bernoulli distribution with p=0.5 can be used to model the outcome of the experiment. The PMF of the Bernoulli distribution with p=0.5 is:

P(X=1) = 0.5 (the probability of success)
P(X=0) = 0.5 (the probability of failure)

The difference between the Bernoulli distribution and the binomial distribution is that the Bernoulli distribution describes the outcome of a single binary experiment, while the binomial distribution describes the outcome of multiple independent and identically distributed binary experiments.

The binomial distribution is obtained by adding up the results of n independent and identically distributed Bernoulli trials, where each trial has a probability of success p. The binomial distribution has two parameters, n and p. The probability mass function (PMF) of the binomial distribution is given by:

P(X=k) = (n choose k) p^k (1-p)^(n-k), for k = 0, 1, 2, ..., n

where X is the random variable that represents the number of successes in n independent Bernoulli trials.

In summary, the Bernoulli distribution describes the outcome of a single binary experiment, while the binomial distribution describes the outcome of multiple independent and identically distributed binary experiments.'''
```

```python
#Q8: What is the z score? State the importance of the z score.
#Ans-
'''The z-score, also known as the standard score, is a measure of how many standard deviations a data point is away from the mean of a distribution. It is calculated by subtracting the population mean from an individual observation and then dividing the result by the population standard deviation.

The formula for the z-score is:

z = (x - μ) / σ

where x is the individual observation, μ is the population mean, and σ is the population standard deviation.

The importance of the z-score is that it allows us to standardize data from different normal distributions and compare them on a common scale. A positive z-score means that the data point is above the mean, while a negative z-score means that the data point is below the mean. A z-score of 0 means that the data point is equal to the mean.

By using the z-score, we can answer questions such as: How unusual is this data point? What proportion of the data falls above or below a certain value? What is the probability of obtaining a value as extreme or more extreme than the one we observed?

The z-score is used in many statistical applications, including hypothesis testing, confidence interval estimation, and process control. It is also a useful tool for identifying outliers in a dataset.'''
```

```python
#Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.
#Ans-
'''The Central Limit Theorem (CLT) is a fundamental concept in probability theory and statistics that states that as the sample size of a random variable increases, the sampling distribution of the sample means approaches a normal distribution, regardless of the shape of the original distribution.

In other words, if we take multiple random samples from a population, the mean of each sample will follow a normal distribution, with the mean of the sampling distribution being equal to the population mean and the standard deviation of the sampling distribution being equal to the population standard deviation divided by the square root of the sample size.

The significance of the Central Limit Theorem is that it allows us to use the normal distribution to make inferences about population parameters, even when the population distribution is not normal. It provides a theoretical foundation for many statistical methods, including hypothesis testing, confidence interval estimation, and process control.

The Central Limit Theorem is particularly important in practical applications because it provides a way to estimate population parameters from sample data, which is often more feasible and less expensive than collecting data on the entire population. 
By using the Central Limit Theorem, we can obtain accurate estimates of population parameters even when the sample size is relatively small or the population distribution is unknown or non-normal.'''
```
